```python
import requests
from bs4 import BeautifulSoup
import os
import time
```

```python
BASE = "https://informacoes.anatel.gov.br"
URL_INICIAL = BASE + "/legislacao/resolucoes"

HEADERS = {"User-Agent": "Mozilla/5.0"}

pasta_base = "resolucoes_anatel"
os.makedirs(pasta_base, exist_ok=True)


def get_soup(url):
    r = requests.get(url, headers=HEADERS)
    r.raise_for_status()
    return BeautifulSoup(r.text, "html.parser")

def coletar_links_anos():
    soup = get_soup(URL_INICIAL)
    links_anos = {}

    for a in soup.find_all("a", href=True):
        href = a["href"]

        if "/legislacao/resolucoes/" in href and href.rstrip("/").split("/")[-1].isdigit():
            ano = href.rstrip("/").split("/")[-1]

            if href.startswith("/"):
                href = BASE + href

            links_anos[ano] = href

    return links_anos

def coletar_resolucoes(url_ano):
    soup = get_soup(url_ano)
    links = []

    for a in soup.find_all("a", href=True):
        href = a["href"]

        if "/legislacao/resolucoes/" in href and not href.rstrip("/").split("/")[-1].isdigit():
            if href.startswith("/"):
                href = BASE + href

            links.append(href)

    return list(set(links))


def salvar_resolucao(url, pasta_ano):
    soup = get_soup(url)

    titulo = soup.title.string.strip()
    if "Resoluções Conjuntas" in titulo:
        return 
    nome_arquivo = titulo + ".html"

    caminho = os.path.join(pasta_ano, nome_arquivo)

    with open(caminho, "w", encoding="utf-8") as f:
        f.write(soup.prettify())

    print("Salvo:", nome_arquivo)

anos = coletar_links_anos()

for ano, link_ano in anos.items():
    print(f"\nProcessando ano {ano}")

    pasta_ano = os.path.join(pasta_base, ano)
    os.makedirs(pasta_ano, exist_ok=True)

    resolucoes = coletar_resolucoes(link_ano)

    for url in resolucoes:
        salvar_resolucao(url, pasta_ano)
```
